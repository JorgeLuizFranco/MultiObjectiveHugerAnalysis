```python
def get_variables_in_col(col):
    return tuple(col.replace("(","").replace(")","").split(", "))

def get_dict_records(table):
    records = {get_variables_in_col(col): {} for col in table.columns if col != "Ind"}
    print("Ok em get_variables_in_col 1")
    for index, row in table.iterrows():
        ind = row["Ind"]
        for el, col in zip(row, table.columns):
            if (col == "Ind"):
                continue
            if el != ".":
                records[get_variables_in_col(col)][ind] = el
                print("Ok em get_variables_in_col 3")
    return records

def get_variables_separated(table):
  old_columns = [col for col in table.columns if col != "Ind"]
  probability_tuples = [get_variables_in_col(col) for col in old_columns]
  print("Ok em get_variables_in_col 4")
  individual_columns = []
  for lista in probability_tuples:
      individual_columns.extend(lista)
  return individual_columns, probability_tuples, old_columns

def get_counting_df(variables):
    pre_dict={}
    name_to_index={}
    for i in range(len(variables)):
        name_to_index[variables[i]]= i

    for i in range(len(variables)):
        pre_dict[variables[i]]=[]
    for i in range(2**len(variables)):       # generating zeros and ones
          valb=str(get_bin(i,len(variables)))
          for i in range(len(valb)):
              pre_dict[variables[i]].append(valb[i])

    print("pre_dict=", pre_dict)
    pdf= pd.DataFrame(pre_dict)
    print("pd.columns=",pdf.columns)
    return pd.DataFrame(pre_dict)

def create_base_df(table):
    individual_columns, probability_tuples, old_columns = get_variables_separated(table)
    print("Ok em get_variables_separated")
    df = get_counting_df(individual_columns)
    print("individual_columns=", individual_columns)
    print("Ok em get_counting_df_out")
    for col in old_columns:
        df[col] = [-1 for _ in range(len(df))]
    return df

def get_probability_df(table):
    records = get_dict_records(table)
    print("Ok em get_dict_records")
    df = create_base_df(table)
    print("Ok em create_base_df")
    len_old_columns = len(df.columns) - len(records)

    for index, row in df.iterrows():
        for header in df.columns[len_old_columns:]:
            variables = get_variables_in_col(header)
            print("Ok em get_variables_in_col 2")
            ind = ""
            for var in variables:
                ind += str(row[var])
            df.loc[index, header] = records[variables][ind]

    rename_mapper = {col : "P" + col for col in df.columns[len_old_columns:]}
    df.rename(columns=rename_mapper, inplace=True)
    return df
```

## Minimum Description Length (MDL) for **MPM**

```python
class MPM:

    def __init__(self, cm:int, cp:int, complexity: int, prob_table: pd.DataFrame):
        self.cm = cm
        self.cp = cp
        self.complexity = complexity
        self.prob_table= prob_table
```

```python
def get_partitions(table):
  tables = []

  for var_name in table.columns:
    if var_name=='Ind':
      continue

    df= table[['Ind',var_name]]

    tables.append(df.dropna())

  return tables # [(A,B),(C),(D)]
```

```python
from math import log2

def get_model_complexity(table, partitions):
  # display(table)

  summation = 0
  for partition in partitions:
      #print('Partição: ')
      #display(partition) #(A,B)
      l_bb_i = len(partition.iloc[0]["Ind"]) # tamanho do bb len(A,B)=2
      #print(l_bb_i)
      summation += 2 ** l_bb_i - 1

  Np = len(table) # tamanho da população
  return log2(Np + 1) * summation
```

```python
from collections import Counter

def create_sorted_tuple_set(arr):
  # Create a Counter object from the input array
  frequency_counter = Counter(arr)

  # Convert the Counter object to a list of tuples
  tuple_list = list(frequency_counter.items())

  # Sort the list of tuples based on the frequency of each element
  sorted_tuple_list = sorted(tuple_list, key=lambda x: x[1], reverse=True)
  return sorted_tuple_list
```

```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

def _draw_as_table(df, pagesize):
    alternating_colors = [['white'] * len(df.columns), ['lightgray'] * len(df.columns)] * len(df)
    alternating_colors = alternating_colors[:len(df)]
    fig, ax = plt.subplots(figsize=pagesize)
    ax.axis('tight')
    ax.axis('off')
    the_table = ax.table(cellText=df.values,
                        rowLabels=df.index,
                        colLabels=df.columns,
                        rowColours=['lightblue']*len(df),
                        colColours=['lightblue']*len(df.columns),
                        cellColours=alternating_colors,
                        loc='center')
    return fig


def dataframe_to_pdf(df, filename, numpages=(1, 1), pagesize=(11, 8.5)):
  with PdfPages(filename) as pdf:
    nh, nv = numpages
    rows_per_page = len(df) // nh
    cols_per_page = len(df.columns) // nv
    for i in range(0, nh):
        for j in range(0, nv):
            page = df.iloc[(i*rows_per_page):min((i+1)*rows_per_page, len(df)),
                           (j*cols_per_page):min((j+1)*cols_per_page, len(df.columns))]
            fig = _draw_as_table(page, pagesize)
            if nh > 1 or nv > 1:
                # Add a part/page number at bottom-center of page
                fig.text(0.5, 0.5/pagesize[0],
                         "Part-{}x{}: Page-{}".format(i+1, j+1, i*nv + j + 1),
                         ha='center', fontsize=8)
            pdf.savefig(fig, bbox_inches='tight')

            plt.close()
```
