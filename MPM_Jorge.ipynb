```python
discrete_dict={}

def one_var_partition(var_name, part):
  number_of_elements= len(df)
  mid_val=-1

  if part=='median':
    mid_val= df[var_name].median()

  elif part=='min_max':
    mid_val= (df[var_name].min()+df[var_name].max())/2

  zero_arr= df[df[var_name]<mid_val][var_name]  # getting values less than mid
  one_arr= df[df[var_name]>=mid_val][var_name] # getting values higher or equal mid

  return zero_arr,one_arr

def two_var_partition(var1, var2, part):
  number_of_elements= len(df) # possible modification
  if part=='median':
    mid_val1= df[var1].median()
    mid_val2= df[var2].median()
  elif part=='min_max':
    mid_val1= (df[var1].min()+df[var1].max())/2
    mid_val2= (df[var2].min()+df[var2].max())/2

  zero_zero_arr= df.loc[(df[var1]<mid_val1) & (df[var2]<mid_val2),[var1,var2]]  # getting values less than mid

  zero_one_arr= df.loc[(df[var1]<mid_val1) & (df[var2]>=mid_val2),[var1,var2]]  # getting values higher or equal mid

  one_zero_arr= df.loc[(df[var1]>=mid_val1) & (df[var2]<mid_val2),[var1,var2]]

  one_one_arr= df.loc[(df[var1]>=mid_val1) & (df[var2]>=mid_val2),[var1,var2]]

  return zero_zero_arr,zero_one_arr,one_zero_arr,one_one_arr

def n_var_partition(vars, part,individuals):
 # (A,B) 00 01 10 11
  mids=[]
  for var in vars:
    if part=='median':
      mids.append(df[var].median())
    elif part=='min_max':
      mids.append((df[var].min()+df[var].max())/2)

  ans={}

  for ind in individuals:

    if ind[0]=='0':
      filt= (df[vars[0]]<mids[0])
    else:
      filt=(df[vars[0]]>=mids[0])

    for i in range(1,len(vars)):
      if ind[i]=='0':
         filt= filt & (df[vars[i]]<mids[i])
      else:
         filt= filt & (df[vars[i]]>=mids[i])

    ans[ind]=len(df[filt])

  return ans
```

```python
def subsets_k(collection, k): yield from partition_k(collection, k, k)
def partition_k(collection, min, k):
  if len(collection) == 1:
    yield [ collection ]
    return

  first = collection[0]
  for smaller in partition_k(collection[1:], min - 1, k):
    if len(smaller) > k: continue
    # insert `first` in each of the subpartition's subsets
    if len(smaller) >= min:
      for n, subset in enumerate(smaller):
        yield smaller[:n] + [[ first ] + subset]  + smaller[n+1:]
    # put `first` in its own subset
    if len(smaller) < k: yield [ [ first ] ] + smaller

variables=['TNat','TVacLac','T_Cres','T_Descarte','TPrenhez','ICDP','Lucro_Bruto','VPUP']

all_partitions=[]

for i in range(1,len(variables)):
  for n, p in enumerate(subsets_k(variables, i), 1):
    all_partitions.append(p)
    print(n,p)
```

## Transforming in the classic way

```python
def get_variables_in_col(col):
    return tuple(col.replace("(","").replace(")","").split(", "))

def get_dict_records(table):
    records = {get_variables_in_col(col): {} for col in table.columns if col != "Ind"}
    for index, row in table.iterrows():
        ind = row["Ind"]
        for el, col in zip(row, table.columns):
            if (col == "Ind"):
                continue
            if el != ".":
                records[get_variables_in_col(col)][ind] = el
    return records

def get_variables_separated(table):
  old_columns = [col for col in table.columns if col != "Ind"]
  probability_tuples = [get_variables_in_col(col) for col in old_columns]
  individual_columns = []
  for lista in probability_tuples:
      individual_columns.extend(lista)
  return individual_columns, probability_tuples, old_columns

def get_counting_df(variables):
    pre_dict={}
    name_to_index={}
    for i in range(len(variables)):
        name_to_index[variables[i]]= i

    for i in range(len(variables)):
        pre_dict[variables[i]]=[]
    for i in range(2**len(variables)):       # generating zeros and ones
          valb=str(get_bin(i,len(variables)))
          for i in range(len(valb)):
              pre_dict[variables[i]].append(valb[i])

    return pd.DataFrame(pre_dict)

def create_base_df(table):
    individual_columns, probability_tuples, old_columns = get_variables_separated(table)
    df = get_counting_df(individual_columns)
    for col in old_columns:
        df[col] = [-1 for _ in range(len(df))]
    return df

def get_probability_df(table):
    records = get_dict_records(table)
    df = create_base_df(table)
    len_old_columns = len(df.columns) - len(records)

    for index, row in df.iterrows():
        for header in df.columns[len_old_columns:]:
            variables = get_variables_in_col(header)
            ind = ""
            for var in variables:
                ind += str(row[var])
            df.loc[index, header] = records[variables][ind]

    rename_mapper = {col : "P" + col for col in df.columns[len_old_columns:]}
    df.rename(columns=rename_mapper, inplace=True)
    return df
```

## Minimum Description Length (MDL) for **MPM**

```python
class MPM:

    def __init__(self, cm:int, cp:int, complexity: int, prob_table: pd.DataFrame):
        self.cm = cm
        self.cp = cp
        self.complexity = complexity
        self.prob_table= prob_table
```

```python
def get_partitions(table):
  tables = []

  for var_name in table.columns:
    if var_name=='Ind':
      continue

    df= table[['Ind',var_name]]

    tables.append(df.dropna())

  return tables # [(A,B),(C),(D)]
```

```python
from math import log2

def get_model_complexity(table, partitions):
  # display(table)

  summation = 0
  for partition in partitions:
      #print('Partição: ')
      #display(partition) #(A,B)
      l_bb_i = len(partition.iloc[0]["Ind"]) # tamanho do bb len(A,B)=2
      #print(l_bb_i)
      summation += 2 ** l_bb_i - 1

  Np = len(table) # tamanho da população
  return log2(Np + 1) * summation
```

```python
def get_entropy(partition):
    e = 0 # (A,B)
    for index, row in partition.iterrows():
        pk = row.iloc[1]  #frequencia da atual instancia no bb [00,01,10,11]
        if pk > 0:
          e -= pk * log2(pk)
    return e

def get_compressed_population_complexity(table, partitions):
    summation = 0
    for partition in partitions:
        summation += get_entropy(partition)  # adicionando a entropia do atual bb
    return len(table) * summation  # tamanho população * somatorio da entropia
```

```python
from collections import Counter

def create_sorted_tuple_set(arr):
  # Create a Counter object from the input array
  frequency_counter = Counter(arr)

  # Convert the Counter object to a list of tuples
  tuple_list = list(frequency_counter.items())

  # Sort the list of tuples based on the frequency of each element
  sorted_tuple_list = sorted(tuple_list, key=lambda x: x[1], reverse=True)
  return sorted_tuple_list
```
